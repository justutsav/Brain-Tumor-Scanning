```python
# Importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Reading and resizing the image
img = cv2.imread('BrainTumor.jpg')

print(f"Original image size: {img.shape}")
print(f"Data type: {img.dtype}")

# Resizing to 400x400
new_img = cv2.resize(img, (400, 400))
print(f"Resized image shape: {new_img.shape}")

# To show resized image
plt.imshow(new_img)
plt.title("Resized Image")
plt.axis("off")
plt.show()

# Step 2: Dividing the image into smaller 5x5x3 blocks
block_size = 5
rows, cols, channels = new_img.shape

# Creating an empty array for the result
result_arr = np.zeros_like(new_img) 

# Loop through each block
for i in range(0, rows, block_size):
    for j in range(0, cols, block_size):
        # Extracting the block
        block = new_img[i:i+block_size, j:j+block_size]

        # Here later we can calculate entropy  for now, let's just use threshold logic
        # Using your thresholding idea for demonstration
        avg_val = np.mean(block)  # Average brightness of the block
        if avg_val > 190:
            result_arr[i:i+block_size, j:j+block_size] = 255
        else:
            result_arr[i:i+block_size, j:j+block_size] = 0

# Step 3: Display the final result
plt.imshow(result_arr)
plt.title("Processed Image (Block-based Thresholding)")
plt.axis('off')
plt.show()
```
